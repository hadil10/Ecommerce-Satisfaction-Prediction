```python
import pandas as pd

# On utilise des slashs normaux (/) pour éviter les bugs Windows
chemin_data = "C:/Users/Lenovo/Desktop/Projet_IA_BI_Ecommerce/data/raw/"

print("Chargement des données en cours...")

# Chargement des 5 fichiers
df_orders = pd.read_csv(chemin_data + "olist_orders_dataset.csv")
df_reviews = pd.read_csv(chemin_data + "olist_order_reviews_dataset.csv")
df_items = pd.read_csv(chemin_data + "olist_order_items_dataset.csv")
df_customers = pd.read_csv(chemin_data + "olist_customers_dataset.csv")
df_products = pd.read_csv(chemin_data + "olist_products_dataset.csv")

print("✅ Données chargées avec succès !")
print(f"Nombre de commandes : {df_orders.shape[0]}")
print(f"Nombre d'avis clients : {df_reviews.shape[0]}")
```

```python
print("Début de la fusion des tables...")

# 1. On relie les Commandes avec les Clients (grâce à la colonne 'customer_id')
df_master = pd.merge(df_orders, df_customers, on='customer_id', how='inner')

# 2. On ajoute les Avis clients (grâce à la colonne 'order_id')
df_master = pd.merge(df_master, df_reviews, on='order_id', how='inner')

# 3. On ajoute les Articles de la commande (grâce à la colonne 'order_id')
df_master = pd.merge(df_master, df_items, on='order_id', how='inner')

# 4. On ajoute les détails des Produits (grâce à la colonne 'product_id')
df_master = pd.merge(df_master, df_products, on='product_id', how='inner')

print(f"✅ Fusion terminée ! Le tableau final contient {df_master.shape[0]} lignes et {df_master.shape[1]} colonnes.")

# Afficher les 5 premières lignes pour vérifier le résultat
df_master.head()
```

```python
# Chemin vers le dossier 'processed'
chemin_processed = "C:\Users\Lenovo\Desktop\Projet_IA_BI_Ecommerce\data\processed"

# Sauvegarde en fichier CSV (index=False évite de rajouter une colonne de numéros inutile)
df_master.to_csv(chemin_processed + "dataset_fusionne.csv", index=False)

print("✅ Fichier sauvegardé avec succès dans data/processed/dataset_fusionne.csv")
```

```python
print("Nettoyage des données en cours...")

# 1. On supprime les commandes non livrées (on garde uniquement celles avec une date de livraison)
df_master = df_master.dropna(subset=['order_delivered_customer_date'])

# 2. On remplace les commentaires vides par "Aucun"
df_master['review_comment_title'] = df_master['review_comment_title'].fillna('Aucun')
df_master['review_comment_message'] = df_master['review_comment_message'].fillna('Aucun')

# 3. On remplace les catégories de produits manquantes par "Inconnu"
df_master['product_category_name'] = df_master['product_category_name'].fillna('Inconnu')

# 4. On supprime les quelques lignes restantes avec des infos produits manquantes (poids, taille...)
df_master = df_master.dropna()

print("✅ Nettoyage terminé !")
print(f"Nouvelle taille du tableau : {df_master.shape[0]} lignes.")
print(f"Nombre total de valeurs manquantes restantes : {df_master.isnull().sum().sum()}")
```

```python
print("Création de nouvelles variables (Feature Engineering)...")

# 1. On transforme le texte en vrai format "Date" pour que Python puisse faire des calculs
colonnes_dates = ['order_purchase_timestamp', 'order_delivered_customer_date', 'order_estimated_delivery_date']
for col in colonnes_dates:
    df_master[col] = pd.to_datetime(df_master[col])

# 2. On calcule le temps de livraison réel (en jours)
df_master['temps_livraison_jours'] = (df_master['order_delivered_customer_date'] - df_master['order_purchase_timestamp']).dt.days

# 3. On crée un indicateur de retard (1 si la commande est en retard, 0 si elle est à l'heure)
df_master['retard_livraison'] = (df_master['order_delivered_customer_date'] > df_master['order_estimated_delivery_date']).astype(int)

print("✅ Nouvelles colonnes créées avec succès !")

# On affiche un petit extrait pour vérifier nos calculs
df_master[['order_purchase_timestamp', 'order_delivered_customer_date', 'temps_livraison_jours', 'retard_livraison']].head()
```

```python
# On sauvegarde le dataset propre et enrichi
df_master.to_csv(chemin_processed + "dataset_propre.csv", index=False)
print("✅ Dataset final sauvegardé sous 'data/processed/dataset_propre.csv'")
```
